# 1. User your parser to collect data

```python
# from HW2
import re
import requests
from bs4 import BeautifulSoup
import time
movieList=['roma_2018','joker_2019','doctor_strange_2016','coco_2017','logan_2017','walk_to_remember','love_actually'
          ,'10_things_i_hate_about_you','ps_i_love_you','saw_ii']
page=100
data=[]
movie_list=[]
for movie in movieList:
    src=False
    for x in range(1,page+1):
        for y in range(1,6):
            try:
                url='https://rottentomatoes.com/m/'+movie+'/reviews?page='+str(x)
                print(url)
                response = requests.get(url)
                src = response.content
                break
            except:
                print ('failed attempt #',y)
                time.sleep(2)
        if not src:
            print('Could not get page: ', x)
            continue
        else:
            print('Successfully got page: ', x)
        soup=BeautifulSoup(src)
        reviews = soup.findAll('div', {'class':re.compile('row review_table_row')})
        #print(len(reviews))
        for review in reviews:
            name='NA'
            rate='NA'
            source='NA'
            text='NA'
        #grab name
            n=review.find('a', {'href':re.compile('critic')})
            if n:
                name=n.text.strip()
        #grab rate
            r=review.find('div',{'class':re.compile('review_icon icon small (.+)')}).attrs['class'][-1]
            if r:
                rate=r
        #grab source
            s=review.find('a',{'href':re.compile('source')})
            if s:
                source=s.find('em').text.strip()
        #grab text
            t=review.find('div',{'class':re.compile('the_review')})
            if t:
                text=t.text.strip()
        #grab date
            d=review.find('div',{'class':re.compile('review-date')})
            if d:
                date=d.text.strip()
            data.append([movie,name,rate,source,text,date])
        src=False
    movie_list.append([data])
    data=[]
        
    with open('Desktop/elif_aydin_movie.txt', mode='w', encoding='utf-8') as f:
        for d in data:
            f.write(d[0] + '\t' + d[1] + '\t' + d[2] + '\t'+ d[3] + '\t'+ d[4] +'\n' +d[5]+'\n')
```

Specifically n-grams

a. n-grams vectorizer with linearSVC model by using different c values

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
X_t=[]
ngramfit=[]
#turning numerical representation
for i in range(0,10):
    stop_words = ['in', 'of', 'at', 'a', 'the']
    # Create a vectorizer that will track text as binary features
    ngram= CountVectorizer(binary=True, ngram_range=(1, 3), stop_words=stop_words)
    # Let the vectorizer learn what tokens exist in the text data
    ngramfit.append(ngram.fit(cleaned_df[i]['Text']))
    # Turn these tokens into a numeric matrix
    X_t.append(ngram.transform(cleaned_df[i]['Text']))
movieList=['roma_2018','joker_2019','doctor_strange_2016','coco_2017','logan_2017','walk_to_remember','love_actually'
          ,'10_things_i_hate_about_you','ps_i_love_you','saw_ii']
data=[]
for i in range(0,10):
    #splitting data
    X_train, X_val, y_train, y_val = train_test_split(
        X_t[i], cleaned_df[i]['Rate'], train_size = 0.8)
    #modeling
    for c in [0.001, 0.005, 0.01, 0.05, 0.1]:
        svm = LinearSVC(C=c)
        svm.fit(X_train, y_train)
        #print ("Accuracy for C=%s: %s" 
           #% (c, accuracy_score(y_val, svm.predict(X_val))))
        data.append([movieList[i],'ngram_vectorizer','linear_SVC',c,accuracy_score(y_val,svm.predict(X_val))])
```
