```python
import pandas as pd
```

```python
df=pd.read_csv('fashion data.csv')
df.head(5)
```

```python
new_list=[]
for i in range(0,len(df)): 
        new_list.append(df["review text"][i].replace("\r"," "))
print(new_list[0])
```

```python
word_list = '$$'.join(new_list)
word_list.replace('$$',' ')
print(word_list)

```

```python
punctuations = ',.?!:;'
no_punct = ""
for char in word_list:
   if char not in punctuations:
       no_punct = no_punct + char
print(no_punct)
```

```python
no_punct = no_punct.replace('\n',' ') 
print(no_punct)
```

```python
tokens = nltk.word_tokenize(no_punct)
type(tokens)
```

```python
#change all tokens into lower case 
words1 = [w.lower() for w in tokens]   #list comprehension 

#only keep text words, no numbers 
words2 = [w for w in words1 if w.isalpha()]

#encode the raw token list by utf-8
#words3 = [w.encode('utf-8') for w in words2]
```

```python
#generate a frequency dictionary for all tokens 
freq = FreqDist(words2)

#sort the frequency list in descending order
sorted_freq = sorted(freq.items(),key = lambda k:k[1], reverse = True)
sorted_freq
```

```python
freq.plot(30)
```

```python
from nltk.corpus import stopwords

stopwords = stopwords.words('english') #use the NLTK stopwords
```

```python
#only keep the words that not in nltk stopwords word list
words_nostopwords = [w for w in words2 if w not in stopwords]
```

```python
#generate a frequency dictionary for all tokens 
freq_nostw = FreqDist(words_nostopwords)

#sort the frequency list in decending order
sorted_freq_nostw = sorted(freq_nostw.items(),key = lambda k:k[1], reverse = True)
sorted_freq_nostw
```

```python
freq_nostw.plot(30)
```

```python
nltk.PorterStemmer?
```

```python
#Use Porter Stemmer 
porter = nltk.PorterStemmer()
stem1 = [porter.stem(w) for w in words_nostopwords]

#Get the frequency distribution 
freq1 = FreqDist(stem1)
#Sort the result
sorted_freq1 = sorted(freq1.items(),key = lambda k: k[1], reverse = True)
sorted_freq1
```

```python
freq1.plot(30)
```

```python
#Steps are the same with the Porter Stemmer
lancaster = nltk.LancasterStemmer()
stem2 = [lancaster.stem(w) for w in words_nostopwords]
#stem2 = [w for w in stem2]
freq2 = FreqDist(stem2)
sorted_freq2 = sorted(freq2.items(),key = lambda k: k[1], reverse = True)
sorted_freq2
```

```python
freq2.plot(30)
```

```python
from nltk.util import ngrams
```

```python
bigrams=list(ngrams(stem2,2))
print(len(bigrams))
bigrams
```

```python
freq2 = FreqDist(bigrams)
sorted_freq2 = sorted(freq2.items(),key = lambda k: k[1], reverse = True)
sorted_freq2
```

```python
freq2.plot(30)
```

```python
trigrams=list(ngrams(stem2,3))
print(len(trigrams))
trigrams
```

```python
freq2 = FreqDist(trigrams)
sorted_freq2 = sorted(freq2.items(),key = lambda k: k[1], reverse = True)
sorted_freq2
```

```python
freq2.plot(30)
```

```python
#pos_tag(tokens)
#Generate a list of POS tags
POS_tags = nltk.pos_tag(tokens) #use unprocessed 'tokens', not 'words'
POS_tags
```

```python
#Generate a frequency distribution 
tag_freq = nltk.FreqDist(POS_tags)
sorted_tag_freq = sorted(tag_freq.items(), key = lambda k:k[1], reverse = True)
sorted_tag_freq
```

```python
tag_freq.plot(30)
```

```python
#Keep tags that start with 'N'
POS_tag_listN = [(word,tag) for (word,tag) in POS_tags if tag.startswith('NNP')]
```
